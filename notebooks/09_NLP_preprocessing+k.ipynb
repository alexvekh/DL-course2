```python

# function to clean text  / Визначимо функцію нормалізації тексту.
def normalize_text(raw_review):

    # Remove html tags
    text = re.sub("<[^>]*>", " ", raw_review) # match <> and everything in between. [^>] - match everything except >

    # Remove emails
    text = re.sub("\S*@\S*[\s]+", " ", text) # match non-whitespace characters, @ and a whitespaces in the end

    # remove links
    text = re.sub("https?:\/\/.*?[\s]+", " ", text) # match http, s - zero or once, //,
                                                    # any char 0-unlimited, whitespaces in the end

     # Convert to lower case, split into individual words
    text = text.lower().split()

    # Replace contractions with their full versions
    text = [contractions.get(word) if word in contractions else word
            for word in text]

    # Re-splitting for the correct stop-words extraction
    text = " ".join(text).split()

    # Remove stop words
    text = [word for word in text if not word in stop_words]

    text = " ".join(text)

    # Remove non-letters
    text = re.sub("[^a-zA-Z' ]", "", text) # match everything except letters and '


    # Stem words. Need to define porter stemmer above
    # text = [stemmer.stem(word) for word in text.split()]

    # Lemmatize words. Need to define lemmatizer above
    doc = nlp(text)
    text = " ".join([token.lemma_ for token in doc if len(token.lemma_) > 1 ])

    # Remove excesive whitespaces
    text = re.sub("[\s]+", " ", text)

    # Join the words back into one string separated by space, and return the result.
    return text
```

# Building a model

```python
# Bag of Words - метод для представлення тексту в обробці природної мови (NLP).
# Це модель представлення тексту, в якій текст (документ або речення) розглядається як невпорядкована сукупність слів,
# тобто «мішок», без урахування граматики й порядку слів, але з урахуванням їх частоти.

# Основні кроки побудови моделі BoW:
  # 1. Збір даних. Отримуємо корпус (corpus) документів (текстів), які потрібно проаналізувати.
  # 2. Токенізація. Розбиваємо кожен документ на окремі слова (токени).
  # 3. Створення словника. Формуємо унікальний список усіх слів, що зустрічаються в усіх документах.
  # 4. Побудова векторів. Для кожного документа створюємо вектор, довжина якого дорівнює кількості слів у словнику. Кожен елемент вектора представляє частоту відповідного слова в документі.

# Розділимо дані на навчальні й тестові.
train_idxs = df.sample(frac=0.8, random_state=42).index
test_idxs = [idx for idx in df.index if idx not in train_idxs]
```

```python
X_train = df.loc[train_idxs, 'text_normalized']
X_test = df.loc[test_idxs, 'text_normalized']

y_train = df.loc[train_idxs, 'sentiment']
y_test = df.loc[test_idxs, 'sentiment']
```

```python
# Creating and training a CountVectorizer object
# Створюємо й навчаємо об'єкт CountVectorizer. Це клас із бібліотеки scikit-learn,
# який перетворює текстові дані в числові вектори, використовуючи підхід «Bag of Words».

vect = CountVectorizer().fit(X_train)
len(vect.vocabulary_)
```

```python
# features examples / Подивимось на приклад ознак, які було виокремлено. Тут ознакою буде кожен унікальний токен словника.
vect.get_feature_names_out()[:5]
```

```python
# transform the documents in the training data to a document-term matrix
# Перетворимо навчальні дані на матрицю документ-термін (document-term matrix).
X_train_vectorized = vect.transform(X_train)
X_train_vectorized.shape
```

```python
# Resulted features representation is a sparse matrix

X_train_vectorized
```

```python
# Представлення отриманих ознак є розрідженою матрицею (sparse matrix).
# Розріджена матриця — це тип матриці, в якій більшість елементів є нулями.
X_train_vectorized[:,0].todense()
```

```python
# Створимо клас моделі й навчимо її на отриманих даних.
model = LogisticRegression(random_state=42)
model.fit(X_train_vectorized, y_train)
```

```python
# Розрахуємо точність моделі.
predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
# Для подальшого швидкого тестування різних варіантів обробки даних напишемо допоміжну функцію.
# Алгоритм tf-idf ми розглянемо в наступному підрозділі.

def get_preds(text_column, algorithm, ngrams=(1,1)):

    X_train = df.loc[train_idxs, text_column]
    X_test = df.loc[test_idxs, text_column]

    y_train = df.loc[train_idxs, 'sentiment']
    y_test = df.loc[test_idxs, 'sentiment']

    if algorithm == 'cv':
        vect = CountVectorizer(ngram_range=ngrams).fit(X_train)
    elif algorithm == 'tfidf':
        vect = TfidfVectorizer(ngram_range=ngrams).fit(X_train)
    else:
        raise ValueError('Select correct algorithm: `cv` or `tfidf`')

    print('Vocabulary length: ', len(vect.vocabulary_))

    # transform the documents in the training data to a document-term matrix

    X_train_vectorized = vect.transform(X_train)
    print('Document-term matrix shape:', X_train_vectorized.shape)

    model = LogisticRegression(random_state=42)
    model.fit(X_train_vectorized, y_train)

    predictions = model.predict(vect.transform(X_test))

    print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
# Перевіримо результати моделювання на нетокенізованих даних.
get_preds('Text', 'cv')
# Бачимо трохи кращі значення метрики AUC на необроблених даних.
# Можемо зробити висновок, що під час попередньої обробки певна інформація була втрачена.
```

# TF-IDF

Term frequency-inverse document frequency (tf-idf) — це статистичний показник, який оцінює релевантність слова документа в колекції документів. Це робиться шляхом множення двох показників: скільки разів слово з’являється в документі (tf) та зворотної частоти цього слова в наборі документів (idf).

Періодичність терміна (tf). У документі d періодичність представляє кількість випадків, коли з’являється слово t, поділену на кількість слів у d.


Частота документів (df) вимірює важливість документів у цілому корпусі.  Це дуже схоже на tf, але єдина відмінність полягає в тому, що tf є лічильником частоти для терміна t у документі d, тоді як df є кількістю входжень терміна t у наборі документів N.


Зворотна частота документа (idf) вимірює інформативність терміна t. Коли ми обчислюємо idf, він буде дуже низьким для слів, які найчастіше зустрічаються, наприклад, для стоп-слів (оскільки вони присутні майже в усіх документах, і N/df ​

```python
# Для ненормалізованих даних:
get_preds('Text', 'tfidf')
```

Спостерігаємо, що результати на необроблених даних з використанням алгоритму BoW поки що є найкращими.

Дійсно, TF-IDF враховує частоту вживання термінів у всьому корпусі, що може призвести до підсилення ваги неважливих слів або шуму в тексті. Тоді як bag of words просто рахує кількість входжень слів у кожному документі й може бути менш чутливим до такого шуму, а значить, і показувати кращі результати.

# N-Grams

N-грами (n-grams) — це безперервні послідовності слів, символів або токенів у документі. З технічної точки зору їх можна визначити як сусідні послідовності елементів у документі.

Приклад

Дано речення: «I live in Ukraine».
- Uni-gram	[«I», «live», «in», «Ukraine»]
- Bi-gram	[«I live», «live in», «in Ukraine»]
- Tri-gram	[«I live in», «live in Ukraine»]

N-грами часто використовуються для створення моделей мови, які прогнозують наступне слово або символ у тексті на основі попередніх N-1 слів або символів.
